## **Import library**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## **Import dataset**

```python
dataset = pd.read_csv('data.csv')
```

```python
print(dataset.head())
```

```python
# Memisahkan fitur (X) dan target (y)
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
print(dataset.info())
```

```python
# periksa nilai yang hilang di setiap kolom
print(dataset.isnull().sum())
```

Dataset ini tidak memiliki nilai yang hilang. Semua kolom memiliki data yang lengkap, sehingga dapat melanjutkan analisis dan pemodelan lebih lanjut tanpa perlu mengisi nilai yang hilang.

## **Encoding data kategori (Atribut)**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
#ubah kolom Sex bertipe string ke nuemrik
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough') 
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

## **Encoding data kategori (Label)**

```python
from sklearn.preprocessing import LabelEncoder
# Encoding target "Anaemic" menggunakan LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

## **Membagi dataset ke dalam training set dan test set**

```python
from sklearn.model_selection import train_test_split

# test_size = 0.2 berarti 20% dari data akan digunakan untuk pengujian
# random_state = 0 untuk memastikan hasil yang konsisten setiap kali kode dijalankan
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

## **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
# Menerapkan standarisasi pada fitur numerik (mulai dari kolom ke-5)
scaler = StandardScaler()
X_train[:, 4:] = scaler.fit_transform(X_train[:, 4:])
X_test[:, 4:] = scaler.transform(X_test[:,  4:])
```

```python
print(X_train)
```
